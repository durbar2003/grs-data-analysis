```python
import requests
import io
import numpy as np
import matplotlib.pyplot as plt

# Define the URL for the data
url = 'https://pds-geosciences.wustl.edu/lunar/lp-l-grs-3-rdr-v1/lp_2xxx/grs/1998_270_grs.dat'

# Retrieve the data from the URL
response = requests.get(url)

# Load the data from the response content
data = np.frombuffer(response.content, dtype=np.float32)

# Plot the data
plt.plot(data)
plt.xlabel('Channel')
plt.ylabel('Counts')
plt.show()
```
